# Employee Attrition

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
style.use("ggplot")

from numpy.random import rand
from numpy.random import seed

pd.set_option('display.max_columns', None)
```

## The Data

This sample data set is provided by IBM through the Watson Analytics Sample Data, it can be obtained following the next URL:

https://www.ibm.com/communities/analytics/watson-analytics-blog/hr-employee-attrition/

Last accessed on 03/28/2019

```python
employeesData = pd.read_csv('employeeAttrition.csv')
employees = employeesData
```

```python
display(employees)
```

## Data Cleaning & Preparation

```python
display(employees.describe())
```

From this output we can see there are some variables that can be removed as they don't give useful information:  

{Over18, StandardHours, EmployeeCount, EmployeeNumber}

```python
employees = employees.drop(['Over18', 'StandardHours', 'EmployeeCount', 'EmployeeNumber'], axis=1)
display(employees)
```

There are other variables we couldn't find their meaning so we're also removing them:

{HourlyRate, MonthlyRate, DailyRate}

```python
employees = employees.drop(['HourlyRate', 'MonthlyRate', 'DailyRate'], axis=1)
display(employees)
```

```python
employees = employees.astype({'WorkLifeBalance': 'category'
                  , 'JobSatisfaction': 'category'
                  , 'StockOptionLevel': 'category'
                  , 'RelationshipSatisfaction': 'category'
                  , 'PerformanceRating': 'category'
                  , 'JobInvolvement': 'category'
                  , 'EnvironmentSatisfaction': 'category'
                  , 'Education': 'category'
                  , 'EducationField': 'category'
                 })
```

Next we set the numeric type to the corresponding variables and produce the dummy variables for the nominal ones to ease our further efforts while building the model. 

```python
travel = {'Travel_Rarely': 1, 'Travel_Frequently': 2, 'Non-Travel': 3}
travelOrd = employees['BusinessTravel'].map(travel)
employees.BusinessTravel = travelOrd

d = {'Yes': 1, 'No': 0}
OverTimeBool = employees['OverTime'].map(d)
employees.OverTime = OverTimeBool

AttritionBool = employees['Attrition'].map(d)
employees.Attrition = AttritionBool

employees = pd.get_dummies(employees, 
                                columns=["Department", "EducationField", "JobRole", "MaritalStatus", "Gender"], 
                                prefix=["Department", "EducationField", "JobRole", "MaritalStatus", "Gender"])

employees["Education"] = pd.to_numeric(employees["Education"])
employees["EnvironmentSatisfaction"] = pd.to_numeric(employees["EnvironmentSatisfaction"])
employees["JobInvolvement"] = pd.to_numeric(employees["JobInvolvement"])
employees["JobSatisfaction"] = pd.to_numeric(employees["JobSatisfaction"])
employees["PerformanceRating"] = pd.to_numeric(employees["PerformanceRating"])
employees["RelationshipSatisfaction"] = pd.to_numeric(employees["RelationshipSatisfaction"])
employees["StockOptionLevel"] = pd.to_numeric(employees["StockOptionLevel"])
employees["WorkLifeBalance"] = pd.to_numeric(employees["WorkLifeBalance"])
```

### Nominal Variables

First we check the relationship between the independent nominal variables and the dependent variable, for this we execute a Spearman Correlation Test for each variable in this section:

```python
employeeNom = employees[['Attrition' 
                         , 'Department_Human Resources'
                         , 'Department_Research & Development'
                         , 'Department_Sales'
                         , 'EducationField_Human Resources'
                         , 'EducationField_Life Sciences'
                         , 'EducationField_Marketing'
                         , 'EducationField_Medical'
                         , 'EducationField_Other'
                         , 'EducationField_Technical Degree'
                         , 'Gender_Female'
                         , 'Gender_Male'
                         , 'JobRole_Healthcare Representative'
                         , 'JobRole_Human Resources'
                         , 'JobRole_Laboratory Technician'
                         , 'JobRole_Manager'
                         , 'JobRole_Manufacturing Director'
                         , 'JobRole_Research Director'
                         , 'JobRole_Research Scientist'
                         , 'JobRole_Sales Executive'
                         , 'JobRole_Sales Representative'
                         , 'MaritalStatus_Divorced'
                         , 'MaritalStatus_Married'
                         , 'MaritalStatus_Single'                       
                         , 'OverTime']]
display(employeeNom)
```

```python
from scipy.stats import chi2_contingency
from scipy.stats import chi2
from scipy.stats import spearmanr
from scipy.stats import kendalltau
```

The impact of each independent variable does not seem to be strong in most cases, however, there are correlations present between these and the dependent variable Attrition.

```python
list_of_tuples = list(zip(employeeNom.columns))
chiNoms = pd.DataFrame(list_of_tuples, columns=['cols'])
spearNoms = pd.DataFrame(list_of_tuples, columns=['cols'])

plot_cols = 5
plot_rows = len(list_of_tuples) / plot_cols

fig = plt.figure(figsize=(20, 20))
for index, row in chiNoms.iterrows():
    #print('\n*** ', row['cols'], row['pvalue'])
    c_table = pd.crosstab(employeeNom.Attrition, employeeNom[row['cols']])
    stat, p_chi, dof, expected = chi2_contingency(c_table)
    chiNoms.loc[index,'ChiSquared'] = p_chi
    
    coef, p_spear = spearmanr(employeeNom.Attrition, employeeNom[row['cols']])
    spearNoms.loc[index,'spear'] = p_spear
    
    table=pd.crosstab(employeeNom[row['cols']], employeeNom.Attrition)
    table_div = table.div(table.sum(axis=1).astype(float), axis=0)
    ax = fig.add_subplot(plot_rows, plot_cols, index+1)
    table_div.plot(kind='bar', stacked=True, ax=ax)
```

```python
#plt.show()
'''
display(chiNoms)

y_pos = np.arange(len(chiNoms.cols))
plt.figure(figsize=(20,20))

plt.subplot(1, 2, 1)
plt.bar(y_pos, chiNoms.ChiSquared, align='center', alpha=0.5)
plt.xticks(y_pos, chiNoms.cols, rotation='vertical')
plt.ylabel('P-Values')
plt.title('Chi Squared test')
plt.plot([0., 25], [0.05, 0.05], "k--")
'''

display(spearNoms)

drops = pd.DataFrame(spearNoms[spearNoms.spear >= 0.05].cols)
display(drops)

y_pos = np.arange(len(spearNoms.cols))
plt.figure(figsize=(20,10))

plt.subplot(1, 2, 2)
plt.bar(y_pos, spearNoms.spear, align='center', alpha=0.5)
plt.xticks(y_pos, spearNoms.cols, rotation='vertical')
plt.ylabel('P-Values')
plt.title('Spearman test')
plt.plot([0., 25], [0.05, 0.05], "k--")
#plt.show()
```

From this we can see which deparments, education fields and job roles show a weaker correlation to Attrition and consider removing those from our modeling phase.


### Ordinal Variables

First we'll try assuming there's an underlying numerical value in each ordinal variable, also we can assume that the distance between each ordinal value is the same, we'll support this assuptions looking at a best fit line for each of this variables:

```python
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
employeeOrd = employees[['Attrition', 'BusinessTravel', 'Education', 'EnvironmentSatisfaction', 'JobInvolvement', 'JobLevel', 'PerformanceRating', 'RelationshipSatisfaction', 'StockOptionLevel', 'WorkLifeBalance']]
#display(employeeOrd)

list_of_tuples = list(zip(employeeOrd.columns))
ords = pd.DataFrame(list_of_tuples, columns=['cols'])

#employees['BusinessTravel'].unique()

#travel = {'Travel_Rarely': 1, 'Travel_Frequently': 2, 'Non-Travel': 3}
#travelOrd = employeeOrd['BusinessTravel'].map(travel)
#employeeOrd.BusinessTravel = travelOrd

#d = {'Yes': 1, 'No': 0}
#attritionBool = employeeOrd['Attrition'].map(d)
#employeeOrd.Attrition = attritionBool

#arr_y = np.array(attritionBool).reshape(-1, 1)

display(employeeOrd)

# Create linear regression object
regr = linear_model.LinearRegression()
#plt.figure(figsize=(100,50))
plt.subplots(2, 5, figsize=(20, 10))
for index, row in ords.iterrows():
    # Train the model using the training sets
    arr_x = np.array(employeeOrd[row['cols']]).reshape(-1, 1)
    regr.fit(arr_x, employeeOrd['Attrition'])

    # Make predictions using the testing set
    y_pred = regr.predict(arr_x)
    
    # Plot outputs
    plt.subplot(2, 5, index+1)
    plt.scatter(arr_x, employeeOrd['Attrition'],  color='black', s=2)
    plt.plot(arr_x, y_pred, color='blue', linewidth=1)
    plt.title(row['cols'])

    plt.xticks(())
    plt.yticks(())

plt.show()

fig = plt.figure(figsize=(20, 10))
for index, row in ords.iterrows():
    table=pd.crosstab(employeeOrd[row['cols']], employeeOrd.Attrition)
    table_div = table.div(table.sum(axis=1).astype(float), axis=0)
    ax = fig.add_subplot(2, 5, index+1)
    table_div.plot(kind='bar', stacked=True, ax=ax)    
```

We can use Spearma's Rank Correlation method to quantify the assosiation between the indepentend ordinal variables with the dependent variable, as they seem to have a linear relationship with *Attrition*:

```python
list_of_tuples = list(zip(employeeOrd.columns))
#chiNoms = pd.DataFrame(list_of_tuples, columns=['cols', 'pvalue'])
ords = pd.DataFrame(list_of_tuples, columns=['cols'])

for index, row in ords.iterrows():
    coef, p = kendalltau(employeeOrd.Attrition, employeeOrd[row['cols']])
    ords.loc[index,'kendalltau'] = p
    '''    print('Kendall correlation coefficient: %.3f' % coef)
    # interpret the significance
    alpha = 0.05
    if p > alpha:
        print('Samples are uncorrelated (fail to reject H0) p=%.3f' % p)
    else:
        print('Samples are correlated (reject H0) p=%.3f' % p)'''

    coef, p = spearmanr(employeeOrd.Attrition, employeeOrd[row['cols']])
    ords.loc[index,'spear'] = p
    # interpret p-value
    '''print('Spearmans correlation coefficient: %.3f' % coef)
    alpha = 0.05
    if p > alpha:
        print('Samples are uncorrelated (fail to reject H0) p=%.3f' % p)
    else:
        print('Samples are correlated (reject H0) p=%.3f' % p)'''

y_pos = np.arange(len(ords.cols))

plt.subplots(1, 2, figsize=(20, 10))

plt.subplot(1, 2, 1)
plt.bar(y_pos, ords.spear, align='center', alpha=0.5)
plt.xticks(y_pos, ords.cols, rotation='vertical')
plt.ylabel('P-Values')
plt.title('Spearman Rank Correlation Test')
plt.plot([0., 9.5], [0.05, 0.05], "k--")

plt.subplot(1, 2, 2)
plt.bar(y_pos, ords.kendalltau, align='center', alpha=0.5)
plt.xticks(y_pos, ords.cols, rotation='vertical')
plt.ylabel('P-Values')
plt.title('Kendallâ€™s Rank Correlation Test')
plt.plot([0., 9.5], [0.05, 0.05], "k--")
plt.show()

display(ords)

ordDrops = pd.DataFrame(ords[ords.spear >= 0.05].cols)

drops = drops.append(ordDrops, ignore_index = True)

display(drops)
```

These tests' p-values show that BusisnessTravel, Education, PerformanceRating and RelationshipSatisfaction don't seem to have a strong relation to the Attrition, supporting what we see in the plots for each variable against Attrition.

## Model Building

Now we'll evaluate some models to find a good fit. We start by setting our seed and splitting the data into training (70%) and test (30%) sets.

```python
from sklearn.model_selection import train_test_split

employeesModel = employees.drop(drops.cols, axis=1)

seed = 42 #(random_state param)

X_train, X_test, y_train, y_test = train_test_split(employeesModel.drop('Attrition', axis=1), employeesModel.Attrition, test_size=0.3, random_state=seed)

display(X_train.describe())

#pd.crosstab(employeesModel.Attrition, employeesModel['Gender_Female'])

```


### Logistic Regression

Our first approach is to apply logistic regression:

```python
import statsmodels.api as sm

logit_model = sm.Logit(np.asarray(y_train), np.asarray(X_train))
result = logit_model.fit()

print(result.summary2())
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import confusion_matrix

logreg = LogisticRegression()
logreg.fit(X_train, y_train)

y_pred = logreg.predict(X_test)
y_prob = logreg.predict_proba(X_test)

cmatrix = confusion_matrix(y_test, y_pred)

print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logreg.score(X_train, y_train)))
```

### Evaluating the model

We check the confusion matrix and recall and precision scores:

```python
from sklearn.metrics import confusion_matrix
c_matrix = confusion_matrix(y_test, y_pred)
print(c_matrix)
```

```python
confusion_df = pd.DataFrame(c_matrix
                            , columns=["Predicted Class " + str(class_name) for class_name in [0,1]]
                            , index = ["Class " + str(class_name) for class_name in [0,1]])

display(confusion_df)

from sklearn import metrics

print(metrics.precision_score(y_test, y_pred))
print(metrics.recall_score(y_test, y_pred))
print(metrics.f1_score(y_test, y_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

The recall for the positive class (1 - there is attrition) appears very low, this result might be explained by the imbalance in the class representation, we will try to mitigate this effect using SMOTE:

```python
len(employeesModel.Attrition)
```

```python
from imblearn.over_sampling import SMOTE

sampler = SMOTE(random_state = seed)
X_rs, y_rs = sampler.fit_sample(employeesModel.drop('Attrition', axis=1), employeesModel.Attrition)

columns_x = employeesModel.drop('Attrition', axis=1).columns

y_rs = pd.DataFrame(y_rs, columns=['Attrition'])
X_rs = pd.DataFrame(X_rs, columns = columns_x)

# we can Check the numbers of our data
print("length of oversampled data is ",len(X_rs))
print("Number of no subscription in oversampled data",len(y_rs[y_rs['Attrition']==0]))
print("Number of subscription",len(y_rs[y_rs['Attrition']==1]))
print("Proportion of no subscription data in oversampled data is ",len(y_rs[y_rs['Attrition']==0])/len(X_rs))
print("Proportion of subscription data in oversampled data is ",len(y_rs[y_rs['Attrition']==1])/len(X_rs))


display(pd.DataFrame(X_rs, columns = columns_x))
```

```python
Xrs_train, Xrs_test, yrs_train, yrs_test = train_test_split(employeesModel.drop('Attrition', axis=1), employeesModel.Attrition, test_size=0.3, random_state=seed)

from imblearn.over_sampling import SMOTE

sampler = SMOTE(random_state = seed)
Xrs_train, yrs_train = sampler.fit_sample(Xrs_train, yrs_train)

columns_x = employeesModel.drop('Attrition', axis=1).columns

yrs_train = pd.DataFrame(yrs_train, columns=['Attrition'])
Xrs_train = pd.DataFrame(Xrs_train, columns = columns_x)

logit_model = sm.Logit(np.asarray(yrs_train), np.asarray(Xrs_train))
result = logit_model.fit()

print(result.summary2())

print("*****")

logreg_rs = LogisticRegression()
logreg_rs.fit(Xrs_train, yrs_train)

y_pred = logreg_rs.predict(Xrs_test)
y_prob = logreg_rs.predict_proba(Xrs_test)

cmatrix_rs = confusion_matrix(yrs_test, y_pred)

print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logreg.score(Xrs_train, yrs_train)))

display(cmatrix_rs)

print(classification_report(yrs_test, y_pred))
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
logit_roc_auc = roc_auc_score(y_test, logreg.predict(X_test))
fpr, tpr, thresholds = roc_curve(y_test, logreg.predict_proba(X_test)[:,1])
plt.figure()
plt.plot(fpr, tpr, label='Logistic Regression (area = %0.2f)' % logit_roc_auc)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.savefig('Log_ROC')
plt.show()
```

```python
logit_roc_auc = roc_auc_score(yrs_test, logreg_rs.predict(Xrs_test))

fpr, tpr, thresholds = roc_curve(yrs_test, logreg_rs.predict_proba(Xrs_test)[:,1])

plt.figure()
plt.plot(fpr, tpr, label='Logistic Regression (area = %0.2f)' % logit_roc_auc)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic - SMOTE')
plt.legend(loc="lower right")
plt.savefig('Log_ROC')
plt.show()
```

We got an improvement in the ability of the model to recognize attrition cases, however, it's still too low as the model only recognizes arount 49% of attrition cases. From this we'll try other models and evaluate which gives the best results.
