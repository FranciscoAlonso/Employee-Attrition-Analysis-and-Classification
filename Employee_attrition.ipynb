### Categorical Variables

First we check the relationship between the independent categorical variables and the dependent variable, for this we execute a Chi Squared Test for each variable in this section:

```python
employeeNom = employees[['Attrition' , 'Department' , 'EducationField' , 'Gender' , 'JobRole' , 'MaritalStatus' , 'OverTime']]
display(employeeNom)
```

```python
from scipy.stats import chi2_contingency
from scipy.stats import chi2
from scipy.stats import spearmanr
from scipy.stats import kendalltau
```

```python
list_of_tuples = list(zip(employeeNom.columns))
chiNoms = pd.DataFrame(list_of_tuples, columns=['cols'])

fig = plt.figure(figsize=(20, 10))
for index, row in chiNoms.iterrows():
    #print('\n*** ', row['cols'], row['pvalue'])
    c_table = pd.crosstab(employeeNom.Attrition, employeeNom[row['cols']])
    stat, p, dof, expected = chi2_contingency(c_table)
    chiNoms.loc[index,'ChiSquared'] = p
    
    table=pd.crosstab(employeeNom[row['cols']], employeeNom.Attrition)
    table_div = table.div(table.sum(axis=1).astype(float), axis=0)
    ax = fig.add_subplot(2, 5, index+1)
    table_div.plot(kind='bar', stacked=True, ax=ax)
    
    

#plt.show()
display(chiNoms)

y_pos = np.arange(len(chiNoms.cols))
plt.figure(figsize=(20,10))

plt.subplot(2, 3, 1)
plt.bar(y_pos, chiNoms.ChiSquared, align='center', alpha=0.5)
plt.xticks(y_pos, chiNoms.cols, rotation='vertical')
plt.ylabel('P-Values')
plt.title('Chi Squared test')
plt.plot([0., 4.5], [0.05, 0.05], "k--")
#plt.show()
```

Variable Gender appears to have no correlation to Attrition 


### Ordinal Variables

First we'll try assuming there's an underlying numerical value in each ordinal variable, also we can assume that the distance between each ordinal value is the same, we'll support this assuptions looking at a best fit line for each of this variables:

```python
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
employeeOrd = employees[['Attrition', 'BusinessTravel', 'Education', 'EnvironmentSatisfaction', 'JobInvolvement', 'JobLevel', 'PerformanceRating', 'RelationshipSatisfaction', 'StockOptionLevel', 'WorkLifeBalance']]
#display(employeeOrd)

list_of_tuples = list(zip(employeeOrd.columns))
ords = pd.DataFrame(list_of_tuples, columns=['cols'])

employees['BusinessTravel'].unique()

travel = {'Travel_Rarely': 1, 'Travel_Frequently': 2, 'Non-Travel': 3}
travelOrd = employeeOrd['BusinessTravel'].map(travel)
employeeOrd.BusinessTravel = travelOrd

d = {'Yes': 1, 'No': 0}
attritionBool = employeeOrd['Attrition'].map(d)
employeeOrd.Attrition = attritionBool

#arr_y = np.array(attritionBool).reshape(-1, 1)

display(employeeOrd)

# Create linear regression object
regr = linear_model.LinearRegression()
#plt.figure(figsize=(100,50))
plt.subplots(2, 5, figsize=(20, 10))
for index, row in ords.iterrows():
    # Train the model using the training sets
    arr_x = np.array(employeeOrd[row['cols']]).reshape(-1, 1)
    regr.fit(arr_x, employeeOrd['Attrition'])

    # Make predictions using the testing set
    y_pred = regr.predict(arr_x)
    
    # Plot outputs
    plt.subplot(2, 5, index+1)
    plt.scatter(arr_x, employeeOrd['Attrition'],  color='black', s=2)
    plt.plot(arr_x, y_pred, color='blue', linewidth=1)
    plt.title(row['cols'])

    plt.xticks(())
    plt.yticks(())

plt.show()

fig = plt.figure(figsize=(20, 10))
for index, row in ords.iterrows():
    table=pd.crosstab(employeeOrd[row['cols']], employeeOrd.Attrition)
    table_div = table.div(table.sum(axis=1).astype(float), axis=0)
    ax = fig.add_subplot(2, 5, index+1)
    table_div.plot(kind='bar', stacked=True, ax=ax)    
```

We can use Spearma's Rank Correlation method to quantify the assosiation between the indepentend ordinal variables with the dependent variable, as they seem to have a linear relationship with *Attrition*:

```python
from scipy.stats import spearmanr
from scipy.stats import kendalltau
```

```python
list_of_tuples = list(zip(employeeOrd.columns))
#chiNoms = pd.DataFrame(list_of_tuples, columns=['cols', 'pvalue'])
ords = pd.DataFrame(list_of_tuples, columns=['cols'])

for index, row in ords.iterrows():
    coef, p = kendalltau(employeeOrd.Attrition, employeeOrd[row['cols']])
    ords.loc[index,'kendalltau'] = p
    '''    print('Kendall correlation coefficient: %.3f' % coef)
    # interpret the significance
    alpha = 0.05
    if p > alpha:
        print('Samples are uncorrelated (fail to reject H0) p=%.3f' % p)
    else:
        print('Samples are correlated (reject H0) p=%.3f' % p)'''

    coef, p = spearmanr(employeeOrd.Attrition, employeeOrd[row['cols']])
    ords.loc[index,'spear'] = p
    # interpret p-value
    '''print('Spearmans correlation coefficient: %.3f' % coef)
    alpha = 0.05
    if p > alpha:
        print('Samples are uncorrelated (fail to reject H0) p=%.3f' % p)
    else:
        print('Samples are correlated (reject H0) p=%.3f' % p)'''

y_pos = np.arange(len(ords.cols))

plt.subplots(1, 2, figsize=(20, 10))

plt.subplot(1, 2, 1)
plt.bar(y_pos, ords.spear, align='center', alpha=0.5)
plt.xticks(y_pos, ords.cols, rotation='vertical')
plt.ylabel('P-Values')
plt.title('Spearman Rank Correlation Test')
plt.plot([0., 9.5], [0.05, 0.05], "k--")

plt.subplot(1, 2, 2)
plt.bar(y_pos, ords.kendalltau, align='center', alpha=0.5)
plt.xticks(y_pos, ords.cols, rotation='vertical')
plt.ylabel('P-Values')
plt.title('Kendall’s Rank Correlation Test')
plt.plot([0., 9.5], [0.05, 0.05], "k--")
plt.show()

display(ords)
```

These tests' p-values show that BusisnessTravel, Education, PerformanceRating and RelationshipSatisfaction don't seem to have a strong relation to the Attrition

## Data Cleaning & Preparation
